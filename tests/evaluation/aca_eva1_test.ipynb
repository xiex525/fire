```python
import pandas as pd
from firefin.evaluation.academia.aca_evaluator import AcaEvaluator
```

```python
from firefin.data.fake import gen_df
```

```python
#test sorting
```

```python
factor1 = gen_df(10, 100, index="day", mock="rand")
factor2 = gen_df(10, 100, index="day", mock="rand")

# 1, 2, 3, 4, 5 periods
forward_returns = { i: gen_df(10, 100, index="day", mock="return") for i in range(1, 6) }
market_cap = gen_df(10, 100, index="day", mock="volume")
```

```python
evaluator1 = AcaEvaluator(factor=factor1, forward_returns=forward_returns)
```

```python
#single sort
```

```python
result, stats = evaluator1.run_single_sort(quantiles=5, return_stats=True, market_cap=market_cap)
print(result)
print(stats)
```

```python
#double sort
```

```python
result = evaluator1.run_double_sort(factor2=factor2, quantiles=(5, 5), dependent=False, value_weighted=True, market_cap=market_cap)
print(result)
```

```python
#test Fama_macbeth
```

```python
factor = gen_df(253, 100, index="day", mock="rand")
return_adj =   gen_df(253, 100, index="day", mock="return")
```

```python
evaluator2 = AcaEvaluator(factor=factor, forward_returns=return_adj)
```

```python
reg_result, reg_stats = evaluator2.run_fama_macbeth(return_adj=return_adj, return_stats=True)
print(reg_result)
print(reg_stats)
```
